## Building an regression model using p-value

```python
from sklearn import linear_model
import scipy.stats as stat

class LogisticRegression_with_p_values:
    def __init__(self, *args, **kwargs):
        self.model = linear_model.LogisticRegression(*args, **kwargs)

    def fit(self, X, y):
        # Ensure data is numeric and y is 1D
        X = np.asarray(X, dtype=np.float64)
        y = np.asarray(y).ravel()

        # Fit the logistic regression model
        self.model.fit(X, y)

        # Compute Fisher information matrix
        denom = (2.0 * (1.0 + np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom, (X.shape[1], 1)).T
        f_ij = np.dot((X / denom).T, X)
        f_ij = f_ij.astype(float)

        # Regularization for numerical stability
        epsilon = 1e-6
        f_ij += np.eye(f_ij.shape[0]) * epsilon

        # Invert to get Cramer-Rao lower bound
        cramer_Rao = np.linalg.inv(f_ij)

        # Clip diagonals to zero to avoid invalid sqrt
        diag_cramer = np.diagonal(cramer_Rao)
        diag_cramer_clipped = np.clip(diag_cramer, a_min=0, a_max=None)
        sigma_estimate = np.sqrt(diag_cramer_clipped)

        # Calculate z-scores and p-values
        z_scores = self.model.coef_[0] / sigma_estimate
        p_values = [stat.norm.sf(abs(x)) * 2 for x in z_scores]

        # Store results
        self.coef_ = self.model.coef_
        self.p_value = p_values
        self.intercept_ = self.model.intercept_

        return self
```

```python
reg = LogisticRegression_with_p_values()
```
